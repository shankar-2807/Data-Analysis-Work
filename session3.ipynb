### Merging

```python
emp = pd.DataFrame({'name': ['Shankar', 'Pragati', 'Archita', 'pratik', 'prashant'],
                   'Age': ['22', '23', '24', '25', '26'],
                   'Salary': [45000, 46000, 47000, 48000, 49000],
                   'DID': [1, 2, 3, 4, 5]})
emp
```

```python
dept = pd.DataFrame({'ID': [1, 2, 3, 4],
                     'Dname':['Software development', 'Data Analyst', 'HR', 'Admin',]})
dept
```

```python
pd.merge(emp, dept, left_on= 'DID', right_on = 'ID', how = 'inner')
```

```python
pd.merge(emp, dept, left_on= 'DID', right_on = 'ID', how = 'outer')
```

```python
pd.merge(emp, dept, left_on = 'DID' , right_on = 'ID', how = 'left')
```

```python
pd.merge(emp, dept, left_on = 'DID' , right_on = 'ID', how = 'right')
```

```python
stud = pd.DataFrame({"Name": ["Shankar", "Pragati", "Pratik", "Dipali"],
                    "Age": ["21", "22", "22", "21"],
                    "Marks": [98, 99, 98, 99],
                    "SID": [101, 102, 103, 104]})
stud
```

```python
deptt = pd.DataFrame({"DID": [101, 102, 103],
                    "Dname": ["SW","HR", "SALES"]})
deptt
```

```python
pd.merge(stud, deptt, left_on= "SID", right_on = "DID", how = "inner")
```

```python
pd.merge(stud, deptt, left_on= "SID", right_on = "DID", how = "outer")
```

```python
pd.merge(stud, deptt, left_on= "SID", right_on = "DID", how = "left")
```

```python
pd.merge(stud, deptt, left_on= "SID", right_on = "DID", how = "right")
```

```python
students = pd.DataFrame({"Name": ["Shankar", "Pragati", "Pratik", "Dipali"],
                        "Age": ["21", "22", "22", "21"],
                        "Marks": [98, 99, 98, 99],
                        "SIDD": [101, 102, 103, 104]})
students
```

```python
department = pd.DataFrame({"Dept_ID":[1, 2, 3],
                          "Dept_Name": ["SW", "HR", "ADMIN"]})
department
```

```python
pd.merge(students, department , left_on= "SIDD", right_on="Dept_ID", how="inner")
```

```python
pd.merge(students, department , left_on= "SIDD", right_on="Dept_ID", how="right")
```

```python
pd.merge(students, department , left_on= "SIDD", right_on="Dept_ID", how="left")
```
